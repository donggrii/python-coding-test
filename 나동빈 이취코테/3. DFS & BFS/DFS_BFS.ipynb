### DFS(Depth-First Search)
- DFS **  **   **    **
- DFS ** ( ) **,     
    1.        
    2.               .         .
    3.   2      
    


```python
def dfs(graph, v, visited):
    visited[v] = True      #    
    print(v, end = ' ')
    
    for i in graph[v]:
        if not visited[i]: # False 
            dfs(graph, i, visited)   #       
```

```python
#      (2 )
graph = [
    [],            #   1  0 index    
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

#      (1 )
visited = [False] * 9  #   1  0 index   9 False  
dfs(graph, 1, visited)
```

### BFS(Breadth-First Search)
- BFS **  ** ,  **    **
- BFS   ** **    
- BFS ** ** ,     
    1.       
    2.                 
    3.   2      
    


```python
from collections import deque

def bfs(graph, start, visited):
    queue = deque([start])   # (Queue)   deque  
    visited[start] = True    #    
    while queue:             #    
        v = queue.popleft()  #     
        print(v, end = ' ')
        
        for i in graph[v]:   #       
            if not visited[i]:
                queue.append(i)
                visited[i] = True
```

```python
#      (2 )
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

#      (1 )
visited = [False] * 9
bfs(graph, 1, visited)
```
